```python
# Define the function to categorize age
def categorize_age(age):
    if age < 2:
        return "0 to 2"
    elif 2 <= age <= 5:
        return "2 to 5"
    elif 5 < age <= 10:
        return "5 to 10"
    elif 10 < age <= 20:
        return "10 to 20"
    elif 20 < age <= 30:
        return "20 to 30"
    else:
        return "more than 30"

# Apply the function to the 'Age' column and create a new column 'Age_categorize'
df['YearsCode_categorize'] = df['Age'].apply(categorize_age)
```
